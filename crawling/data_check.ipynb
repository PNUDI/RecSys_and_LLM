```python
import json
# Keywords 컬럼을 JSON 리스트로 변환
df_final["Keywords"] = df_final["Keywords"].apply(json.loads)


df_final["Keyword_Count"] = df_final["Keywords"].apply(len)
df_final
```

```python
df_final["Text"] = df_final["Title"] + " " + df_final["Description"]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Multi-label binarizer로 Keywords를 One-hot Encoding
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df_final["Keywords"])
texts = [title + " " + desc for title, desc in zip(df_final["Title"], df_final["Description"])]


# 출력 라벨 확인
print(f"Number of keywords: {len(mlb.classes_)}")
print(f"Encoded Labels Shape: {y.shape}")

print(f"Number of Texts: {len(texts)}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델과 토크나이저 로드
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(mlb.classes_))
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class KeywordDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=512):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(label, dtype=torch.float)
        }
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
from tqdm import tqdm  



# 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(texts, y, test_size=0.2, random_state=42)


# 데이터셋 및 DataLoader 생성
train_dataset = KeywordDataset(X_train, y_train, tokenizer)
val_dataset = KeywordDataset(X_val, y_val, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

# 라벨 빈도 계산 및 가중치 생성
label_counts = y.sum(axis=0)
class_weights = 1.0 / label_counts
class_weights_tensor = torch.tensor(class_weights, dtype=torch.float).to("cuda")

# 손실 함수
loss_fn = torch.nn.BCEWithLogitsLoss(pos_weight=class_weights_tensor)

# 옵티마이저 및 학습 설정
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)


# 학습 루프
def train_epoch(model, dataloader, optimizer, loss_fn, device):
    model.train()
    total_loss = 0

    # tqdm으로 데이터로더 래핑
    for batch in tqdm(dataloader, desc="Training", leave=False):
        optimizer.zero_grad()
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        loss = loss_fn(logits, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

# 평가 루프
def eval_epoch(model, dataloader, loss_fn, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []

    # tqdm으로 데이터로더 래핑
    for batch in tqdm(dataloader, desc="Evaluating", leave=False):
        with torch.no_grad():
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            loss = loss_fn(logits, labels)
            total_loss += loss.item()

            preds = torch.sigmoid(logits) > 0.5
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())

    return total_loss / len(dataloader), np.vstack(all_preds), np.vstack(all_labels)

# 학습 및 평가
epochs = 3
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")

    train_loss = train_epoch(model, train_loader, optimizer, loss_fn, device)
    val_loss, val_preds, val_labels = eval_epoch(model, val_loader, loss_fn, device)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Validation Loss: {val_loss:.4f}")
```

```python
def predict(texts):
    model.eval()
    encodings = tokenizer(
        texts,
        truncation=True,
        padding="max_length",
        max_length=512,
        return_tensors="pt"
    )
    with torch.no_grad():
        outputs = model(
            input_ids=encodings["input_ids"],
            attention_mask=encodings["attention_mask"]
        )
    logits = outputs.logits
    probs = torch.sigmoid(logits)  # Multi-label Classification에서는 sigmoid 사용
    predictions = (probs > 0.5).int()  # 0.5 기준으로 라벨 선택
    return predictions

# 예측 예시
sample_texts = ["A superhero saves the city", "A thrilling investigation"]
predicted_labels = predict(sample_texts)

# 예측된 키워드 디코딩
predicted_keywords = mlb.inverse_transform(predicted_labels.numpy())
print(predicted_keywords)
```
