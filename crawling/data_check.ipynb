```python
from sklearn.preprocessing import MultiLabelBinarizer

# Multi-label binarizer로 Keywords를 One-hot Encoding
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(df_final["Keywords"])
texts = df_final['Text']


# 출력 라벨 확인
print(f"Number of keywords: {len(mlb.classes_)}")
print(f"Encoded Labels Shape: {y.shape}")

print(f"Number of Texts: {len(texts)}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 모델과 토크나이저 로드
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(mlb.classes_))
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class KeywordDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=512):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]
        encoding = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(0),
            "attention_mask": encoding["attention_mask"].squeeze(0),
            "labels": torch.tensor(label, dtype=torch.float)
        }
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
from tqdm import tqdm  

label_counts = y.sum(axis=0)

# 최소 빈도 기준 설정
min_frequency = 50

# 자주 등장하는 라벨만 선택
valid_labels = np.where(label_counts >= min_frequency)[0]

# 라벨 필터링
y_filtered = y[:, valid_labels]
label_counts_filtered = label_counts[valid_labels]

# 샘플 필터링: 모든 라벨이 제거된 샘플도 제외
valid_samples = np.any(y_filtered, axis=1)  # 한 샘플이라도 유효한 라벨이 있으면 True
texts = texts.to_numpy()
X_filtered = texts[valid_samples]  # 유효한 샘플만 선택
y_filtered = y_filtered[valid_samples]  # 유효한 샘플에 맞춰 라벨도 필터링

# 필터링 결과 확인
print(f"Original Samples: {len(texts)}")
print(f"Filtered Samples: {len(X_filtered)}")
print(f"Original Labels: {len(label_counts)}")
print(f"Filtered Labels: {len(label_counts_filtered)}")
```

```python

# 데이터 분할
X_train, X_val, y_train, y_val = train_test_split(texts, y, test_size=0.2, random_state=42)


# 데이터셋 및 DataLoader 생성
train_dataset = KeywordDataset(X_train, y_train, tokenizer)
val_dataset = KeywordDataset(X_val, y_val, tokenizer)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)
```

```python
# 라벨 빈도 계산 및 가중치 생성
max_weight = 10  # 가중치 최대값
class_weights_clipped = np.clip(1.0 / label_counts, a_min=None, a_max=max_weight)
class_weights_tensor = torch.tensor(class_weights_clipped, dtype=torch.float).to("cuda")


# 손실 함수
loss_fn = torch.nn.BCEWithLogitsLoss(pos_weight=class_weights_tensor)
```

```python

# 옵티마이저 및 학습 설정
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)


# 학습 루프
def train_epoch(model, dataloader, optimizer, loss_fn, device):
    model.train()
    total_loss = 0

    # tqdm으로 데이터로더 래핑
    for batch in tqdm(dataloader, desc="Training", leave=False):
        optimizer.zero_grad()
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        loss = loss_fn(logits, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(dataloader)

# 평가 루프
def eval_epoch(model, dataloader, loss_fn, device):
    model.eval()
    total_loss = 0
    all_preds = []
    all_labels = []

    # tqdm으로 데이터로더 래핑
    for batch in tqdm(dataloader, desc="Evaluating", leave=False):
        with torch.no_grad():
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            loss = loss_fn(logits, labels)
            total_loss += loss.item()

            preds = torch.sigmoid(logits) > 0.5
            all_preds.append(preds.cpu().numpy())
            all_labels.append(labels.cpu().numpy())

    return total_loss / len(dataloader), np.vstack(all_preds), np.vstack(all_labels)

# 학습 및 평가
epochs = 3
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}/{epochs}")

    train_loss = train_epoch(model, train_loader, optimizer, loss_fn, device)
    val_loss, val_preds, val_labels = eval_epoch(model, val_loader, loss_fn, device)

    print(f"Train Loss: {train_loss:.4f}")
    print(f"Validation Loss: {val_loss:.4f}")
```

```python
def preprocess_texts(texts, tokenizer, max_length=512):
    """
    Preprocess a list of texts using the tokenizer.
    """
    encodings = tokenizer(
        texts,
        truncation=True,
        padding="max_length",
        max_length=max_length,
        return_tensors="pt"
    )
    return encodings

def predict_keywords(model, tokenizer, texts, threshold=0.5, top_k=3, device="cuda"):
    """
    Predict keywords for given texts using the trained model.

    Args:
    - model: Trained model.
    - tokenizer: Tokenizer for preprocessing.
    - texts (list of str): Input texts.
    - threshold (float): Threshold for classification.
    - device (str): Device to run the model on ("cuda" or "cpu").

    Returns:
    - List of predicted keywords for each text.
    """
    model.eval()
    encodings = preprocess_texts(texts, tokenizer)
    
    input_ids = encodings["input_ids"].to(device)
    attention_mask = encodings["attention_mask"].to(device)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        probs = torch.sigmoid(logits)  # Convert logits to probabilities

    # predictions = (probs > threshold).int()  # Apply threshold to get binary predictions
    # return predictions.cpu().numpy()

    # 상위 k개 인덱스 선택
    top_k_indices = torch.topk(probs, top_k, dim=1).indices.cpu().numpy()

    # 기존 MultiLabelBinarizer 형식과 동일한 One-hot Encoding 배열 생성
    predicted_labels = np.zeros_like(probs.cpu().numpy(), dtype=int)

    # 상위 K개 인덱스를 1로 설정 (One-hot Encoding)
    for i, indices in enumerate(top_k_indices):
        predicted_labels[i, indices] = 1

    return predicted_labels  # 기존 방식과 동일한 출력 형태


# 테스트 데이터 준비
test_texts = [
    "Bridgerton The eight close-knit siblings of the Bridgerton family look for love and happiness in London high society."
]


# 키워드 예측
predicted_labels = predict_keywords(model, tokenizer, test_texts, threshold=0.5, device=device)
print(predicted_labels)

# MultiLabelBinarizer로 디코딩
predicted_keywords = mlb.inverse_transform(predicted_labels)

# 결과 출력
for text, keywords in zip(test_texts, predicted_keywords):
    print(f"Text: {text}")
    print(f"Predicted Keywords: {keywords}\n")
```
